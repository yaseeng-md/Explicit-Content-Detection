```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ff_model = pd.read_csv("FF + Transformer.csv")
conv_vit = pd.read_csv("Conv + Transformers.csv")
swin_vit = pd.read_csv("SwinTransformers.csv")
```

# Accuracy Plots

```python
plt.plot(ff_model["accuracy"], label="Train Accuracy")
plt.plot(ff_model["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)  # Add grid lines
plt.title("FF + Transformer Model Accuracy Plot")
plt.show()
```

```python
plt.plot(conv_vit["train_acc"], label="Train Accuracy")
plt.plot(conv_vit["val_acc"], label="Validation Accuracy")
plt.ylim(0.5, max(max(conv_vit["train_acc"]), max(conv_vit["val_acc"])) + 0.05)  # Start from 0
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)  # Add grid lines
plt.title("Conv + Transformer Model Accuracy Plot")
plt.show()

```

```python
plt.plot(swin_vit["train_acc"], label="Train Accuracy")
plt.plot(swin_vit["val_acc"], label="Validation Accuracy")
plt.ylim(0.5, max(max(swin_vit["train_acc"]), max(swin_vit["val_acc"])) + 0.05)  # Start from 0
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)  # Add grid lines
plt.title("Swin ViT Model Accuracy Plot")
plt.show()
```

# Loss PLots

```python
plt.plot(ff_model["loss"], label="Train Loss")
plt.plot(ff_model["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)  # Add grid lines
plt.title("FF + Transformer Model Loss Plot")
plt.show()
```

```python
plt.plot(conv_vit["train_loss"], label="Train Loss")
plt.plot(conv_vit["val_loss"], label="Validation Loss")
plt.ylim(0.4,1)  # Start from 0
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)  # Add grid lines
plt.title("Conv + Transformer Model Loss Plot")
plt.show()
```

```python
plt.plot(swin_vit["train_loss"], label="Train Loss")
plt.plot(swin_vit["val_loss"], label="Validation Loss")
plt.ylim(0,1)  # Start from 0
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)  # Add grid lines
plt.title("Conv + Transformer Model Loss Plot")
plt.show()
```
