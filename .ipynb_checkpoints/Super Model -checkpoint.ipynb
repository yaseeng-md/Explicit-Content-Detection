```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("History/Conv + Transformers.csv")
plt.plot(df["train_acc"], label="Train Accuracy")
plt.plot(df["val_acc"], label="Validation Accuracy")
plt.ylim(0.5, max(max(df["train_acc"]), max(df["val_acc"])) + 0.05)  # Start from 0
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)  # Add grid lines

plt.show()
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("History/Conv + Transformers.csv")
plt.plot(df["train_loss"], label="Train Loss")
plt.plot(df["val_loss"], label="Validation Loss")
plt.ylim(0.4,1)  # Start from 0
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)  # Add grid lines

plt.show()
```

```python
df["val_acc"].max()
```

```python
df = pd.read_csv("History/SwinTransformers.csv")

plt.plot(df["train_acc"],label = "Train Accuracy")
plt.plot(df["val_acc"],label = "Validation Accuracy")
plt.ylim(0.5, max(max(df["train_acc"]), max(df["val_acc"])) + 0.05)  # Start from 0

plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.grid(True)  # Add grid lines

plt.show()
```
